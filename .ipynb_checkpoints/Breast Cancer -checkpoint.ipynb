```python
import pandas as pd
```

```python
dataset=pd.read_csv('breastCancer.csv')
dataset.info()
```

```python
dataset.describe()
```

```python
dataset = dataset.drop(columns=['id'])
```

```python
dataset.describe(include='all')
#count :-> Count the non- NAN For each column/row the number of non-NA/null entries.
#unique :-> The describe method for numeric columns doesn't list the number of unique values
#std :-> didn't get it sample standard deviation 
```

```python
dataset.info()
```

```python
dataset = dataset.drop(dataset[dataset.bare_nucleoli == '?'].index)
# I tried to use the "fillna" to fill the nan vlues from the dataframe with the (0) and with the "ffill" method 
# I tried also the dropna to drop the nan vlues from the dataframe
dataset= dataset.astype({'bare_nucleoli':'int64'}) # Just for fun 
```

```python
dataset.describe(include="all")
```

```python
dataset.info()
```

```python
# Récupérer les features et les variables target
X = dataset.iloc[:,:-1] # select from the first column till the last one
y = dataset['class'] 
```

```python
#Utilisez 60% de la donnée pour l'apprentissage et les 40% restants pour le test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, stratify=y)
# stratify :-> If not None, data is split in a stratified fashion, using this as the class labels.
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

grid_params={
    'n_neighbors':[3,5,7,9,11],
    'metric':['euclidean','manhattan']
}
gs=GridSearchCV(
    KNeighborsClassifier(), 
    grid_params, #Dictionary with parameters names (string) as keys and lists of parameter settings to try as values, or a list of such dictionarie
    verbose=5, #Controls the verbosity: the higher, the more messages we can say the progress of the program
    cv=3,#Cross-validation
    n_jobs=-1 #use all cpus
)
gs_results=gs.fit(X_train, y_train) #didn't understand it how it works
```

```python
gs_results.best_params_
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier(n_neighbors=3,metric='manhattan') # charger le classificateur KNN
knn.fit(X_train, y_train)#entraîner le modèles
y_pred=knn.predict(X_test) #evaluer le modèles
```

```python
from sklearn.metrics import precision_score, recall_score
print('precision = ', precision_score(y_test, y_pred,average='weighted'))
print('recall= ', recall_score(y_test,y_pred,average='weighted'))
```

```python
from sklearn.metrics import  confusion_matrix
confusion_matrix(y_test, y_pred)
```
